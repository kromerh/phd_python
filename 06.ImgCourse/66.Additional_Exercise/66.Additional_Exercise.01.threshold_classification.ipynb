```python
import imageio
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from scipy import interpolate
```

```python
img_path = '/Users/hkromer/02_PhD/08.ETH_courses/02.RadImg/Untitled-9_ai.png'
im = imageio.imread(img_path)
im.shape  # im is a numpy array, rows and cols grayscale values
```

```python
plt.imshow(im, cmap='gray')
plt.axis('off')
plt.title('Raw image')
plt.show()
```

```python
values = im.ravel() # only the grayscale values in the image
values[0:500000] = 0
```

```python
len(values)
```

```python
bins = np.arange(0,255+1,1)
plt.hist(values, bins=bins, density=True)
plt.grid()
plt.title('Raw histogram of image values')
plt.xlabel('Grayscale')
plt.ylabel('Density')
plt.show()
my_hist = plt.hist(values, bins=bins, density=False)
```

```python
bins = np.arange(0,255+1,1)
plt.hist(values, bins=bins, density=False)
plt.grid()
plt.title('Histogram of image values without largest')
xlim = plt.gca().get_xlim()
plt.xlim(xlim[0], 254)
plt.ylim(0, 500000)
plt.xlabel('Grayscale')
plt.ylabel('Density')
plt.show()
```

## Thresholding

```python
y = my_hist[0]  # bin values
x = my_hist[1]  # bin edges
x = x[:-1]  # last entry must be removed since this in only the bin edges
assert len(x) == len(y)

interpC = interpolate.interp1d(x, y, fill_value='extrapolate')
```

### Thresholding by hand

```python
threshold_0 = 100
threshold = [threshold_0]  # arbitrarily chosen threshold
ii = 0
n_sim = 10
for ii in range(0, n_sim):

    class1 = np.where(x < threshold[ii])
    class2 = np.where(x >= threshold[ii])

    # average in group1
    x1 = class1[0]
    y1 = interpC(x1)
    CxGs1 = x1*y1  # counts * grey scales
    m1 = simps(CxGs1, x1)/simps(y1, x1)   # with normalization
    
    # average in group2
    x2 = class2[0]
    y2 = interpC(x2)
    CxGs2 = x2*y2  # counts * grey scales
    m2 = simps(CxGs2, x2)/simps(y2, x2)   # with normalization
    
    new_threshold = (m1+m2)/2

    threshold.append(new_threshold)
    
threshold
```

```python
bins = np.arange(0,255+1,1)
plt.hist(values, bins=bins, density=True)
plt.grid()
plt.title('Histogram of image values with TH')
xlim = plt.gca().get_xlim()
ylim = plt.gca().get_ylim()
plt.plot([threshold_0, threshold_0], [0,1e10], linestyle='--', color='red', label='TH0')
plt.plot([threshold[-1], threshold[-1]], [0,1e10], linestyle='-.', color='green', label='TH')
plt.ylim(ylim)
plt.ylim(0, 0.05)
plt.legend(loc='best')
plt.xlabel('Grayscale')
plt.ylabel('Density')
plt.show()
```

```python
class_values = (values > threshold[-1]).astype(int)*255
```

```python
# Reshape new_pixels as a 2-D array: new_image
new_image = class_values.reshape(im.shape)

# Display the new image with 'gray' color map
plt.title('classified image')
plt.axis('off')
plt.imshow(new_image, cmap='gray')
plt.show()
```
