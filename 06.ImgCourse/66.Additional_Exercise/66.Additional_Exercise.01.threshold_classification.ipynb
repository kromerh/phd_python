## Thresholding

```python
y = my_hist[0]  # bin values
x = my_hist[1]  # bin edges
x = x[:-1]  # last entry must be removed since this in only the bin edges
assert len(x) == len(y)
```

### Thresholding by hand

```python
threshold_0 = 5
threshold = [threshold_0]  # arbitrarily chosen threshold
ii = 0
n_sim = 100
for ii in range(0, n_sim):
    class1 = np.where(x < threshold[ii])
    class2 = np.where(x >= threshold[ii])

    m_1 = np.mean(class1)
    m_2 = np.mean(class2)

    new_threshold = (m_1+m_2)/2

    threshold.append(new_threshold)
    
threshold
```
