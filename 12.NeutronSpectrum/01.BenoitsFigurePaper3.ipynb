```python
import time
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
import datetime
from matplotlib.ticker import AutoMinorLocator
import re

import os
os.environ["PATH"] += os.pathsep + '/usr/local/texlive/2018/bin/x86_64-darwin'
plt.rc('text', usetex=True)
plt.rc('font', weight='bold')
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.rm'] = 'Arial'
matplotlib.rcParams['mathtext.it'] = 'Arial:italic'
matplotlib.rcParams['mathtext.bf'] = 'Arial:bold'
matplotlib.rcParams['mathtext.tt'] = 'Arial'
matplotlib.rcParams['mathtext.cal'] = 'Arial'
matplotlib.rcParams['text.latex.preamble'] = [r'\usepackage{sfmath} \boldmath']

fname = '/Users/hkromer/02_PhD/04_Publiations/01_PaperDrafts/2019/01.CharacterizationPaper/He-3data/one_more_try/Data_paper_Heiko_characterization_90deg_V2.xlsx'

data = pd.read_excel(fname, sheet_name=0, skiprows=3)
datas = []

# column names
cols =     ['E_conv', # energy in keV for convolution
    'spec_conv',
    'E_XP', # energy in keV for experiment
    'spec_XP',
    'E_unconv', # energy in keV for unconvolved
    'spec_unconv']

# 0 degree
data_0 = data.iloc[:,np.arange(1,7,1)]
data_0.columns = cols
datas.append(data_0)

# 90 degree
data_0 = data.iloc[:,np.arange(8,14,1)]
data_0.columns = cols
datas.append(data_0)

# # 180 degree
data_0 = data.iloc[:,np.arange(15,21,1)]
data_0.columns = cols
datas.append(data_0)

fig = plt.figure(figsize=(8, 6))
ax1 = fig.add_subplot(1, 1, 1)
markersize = 80
fs = 20
markers = ['s', 'd', 'o']
colors =['darkred', 'darkblue', 'darkorange']
labels = ['0°', '90°', '180°']


# lines = [(0,(5,1)),(0,(5,5)),(0,(5,10))]
lines = ['dotted','dotted','dotted']
labels = [
    '0°, experiment', '0°, convolved', '0°, unconvolved',
    '90°, experiment', '90°, convolved', '90°, unconvolved',
    '180°, experiment', '180°, convolved', '180°, unconvolved'
]
kk = 0
for jj in np.arange(0,3):
    data = datas[jj]
    
    # experiment
    X = data['E_XP'].dropna()
    X = X / 1000 # MeV
    Y = data['spec_XP'].dropna()
    Y = Y / np.max(Y)
    _ = plt.scatter(X, Y, color=colors[jj], marker=markers[jj], s=markersize, alpha=0.75, label=labels[kk])
    kk = kk + 1 # label


    # convolution
    X = data['E_conv'].dropna()
    X = X / 1000 # MeV
    Y = data['spec_conv'].dropna()
    Y = Y / np.max(Y)
    _1 = plt.plot(X, Y, '-', color=colors[jj], alpha=0.75, label=labels[kk])
    kk = kk + 1 # label
    
    # unconvolved
    X = data['E_unconv'].dropna()
    X = X # MeV
    Y = data['spec_unconv'].dropna()
    Y = Y / np.max(Y)
    _2 = plt.plot(X, Y, linestyle=lines[jj], color=colors[jj], alpha=0.75, label=labels[kk])
    kk = kk + 1    







hs, ls = ax1.get_legend_handles_labels()

leg1 = plt.legend([hs[6],hs[7],hs[8],hs[1],hs[3],hs[5],hs[0],hs[2],hs[4]],[ls[6],ls[7],ls[8],ls[1],ls[3],ls[5],ls[0],ls[2],ls[4]],loc="best", fontsize=fs-4)
# leg1 = plt.legend(loc="best", fontsize=fs-4)

leg1.set_title(r"\textbf{Angle}", prop = {'size': fs-4})
ax1.grid(b=True, which='major', linestyle='-')
ax1.grid(b=True, which='minor', linestyle='--')

plt.xlim(1.2,3.0)


ax1.set_xticks(np.arange(1.0,3.2,0.2))
plt.ylim(0,1.1)

plt.xlabel(r'\textbf{Neutron energy [MeV]}', fontsize=fs)
plt.ylabel(r'\textbf{Normalized neutron spectrum [-]}', fontsize=fs)

ax1.tick_params('x', colors='black', labelsize=fs-2)	
ax1.tick_params('y', colors='black', labelsize=fs-2)	
# x
minor_locator = AutoMinorLocator(2)
ax1.xaxis.set_minor_locator(minor_locator)
# y
minor_locator = AutoMinorLocator(2)

ax1.yaxis.set_minor_locator(minor_locator)    



plt.show()
```
