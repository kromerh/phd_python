```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
fname = '/Users/hkromer/02_PhD/02_Data/01_COMSOL/01_IonOptics/02.current_chamber/05.old_flange/particleData/05.old_flange.004.particleData_0.csv'
```

```python
df = pd.read_csv(fname, skiprows=8, header=None)

# find column headers
c = []
with open(fname, 'r') as myfile:
    for line in myfile:
        if 'Index' in line:
            l = line.rstrip().split(',')
            c.append(l)

myfile.close()
cols = c[0]

df = df.iloc[:,[1,2,3]]
df.columns = ['qx', 'qy', 'qz']
```
