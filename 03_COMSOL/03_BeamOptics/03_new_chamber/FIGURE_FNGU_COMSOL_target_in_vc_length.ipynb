```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import AutoMinorLocator
import os

os.environ["PATH"] += os.pathsep + '/usr/local/texlive/2018/bin/x86_64-darwin'
plt.rc('text', usetex=True)
plt.rc('font', weight='bold')
matplotlib.rcParams['font.weight']= 'bold'
matplotlib.rcParams['mathtext.fontset'] = 'custom'
matplotlib.rcParams['mathtext.rm'] = 'Arial'
matplotlib.rcParams['mathtext.it'] = 'Arial:italic'
matplotlib.rcParams['mathtext.bf'] = 'Arial:bold'
matplotlib.rcParams['mathtext.tt'] = 'Arial'
matplotlib.rcParams['mathtext.cal'] = 'Arial'
matplotlib.rcParams['text.latex.preamble'] = [r'\usepackage{sfmath} \boldmath']
plt.rcParams["axes.labelweight"] = "heavy"
```

```python
folder = '/Users/hkromer/02_PhD/02_Data/01_COMSOL/01_IonOptics/03.new_chamber/02.sweeps.target_pos/only_es/'
data = '02.sweeps.target_pos.010.only_es.csv'

outputfolder = '/Users/hkromer/polybox/Thesis/Chapters/FurtherNeutronGeneratorUpgrades/Figures/COMSOL_target_in_vc_length/'
figname = 'COMSOL_target_in_vc_length'
```

```python
df = pd.read_csv(f'{folder}/{data}')
df['distance'] = df['z_target'] + 75 # distance between target bottom cap and TMP mesh
df.head()
```

```python
# plot for each time
fig, ax1 = plt.subplots(figsize=(10, 7))
fs = 26


plt.plot(df['distance'], df['E_max_kV_mm'], linestyle='-', marker='o', color='darkblue', label="Simulation", markersize=10)
plt.plot([-10,140], [8.0, 8.0], linestyle='-.', color = 'darkred', label="Durand and Shroff (1995)")
plt.xlabel("")

ax1.set_xlabel(r"\textbf{Distance target bottom cap to turbopump [mm]}", fontsize=fs)
ax1.set_ylabel(r"\textbf{Maximum electric field strength}" + '\n' + r"\textbf{[kV/mm]}", fontsize=fs)
ax1.tick_params('x', colors='black', labelsize=fs)
ax1.tick_params(axis='y', labelsize=fs-2)
ax1.set_ylim(6, 9.5)
ax1.set_xlim(10, 140)

# minor ticks x
minor_locator = AutoMinorLocator(2)
ax1.xaxis.set_minor_locator(minor_locator)
# minor ticks y
minor_locator = AutoMinorLocator(2)
ax1.yaxis.set_minor_locator(minor_locator)



leg1 = plt.legend(loc='best', fontsize=fs-4)

# leg1.set_title(r"\textbf{Mesh}", prop = {'size': fs-4})
ax1.grid(b=True, which='major', linestyle='-')
ax1.grid(b=True, which='minor', linestyle='--')


plt.tight_layout()
plt.savefig(f"{outputfolder}{figname}.pdf")
plt.show()
```
