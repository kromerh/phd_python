```python
import numpy as np
import pandas as pd
import pymysql
import sqlalchemy as sql
import matplotlib.pyplot as plt
import matplotlib.dates as md
import seaborn as sns
import re
        
class pumping_down_graph():
    
    
    def plot_graph_VC(self, *fnames):
        
        fig, ax = plt.subplots(figsize=(15,6))

        sns.set()
        sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
        for fname in fnames:
            data = pd.read_csv(fname, skiprows=1, delimiter='\t')
            date = re.findall(r'[^/]+(?=/$|$)', fname)[0]
            date = re.findall(r'(.+).txt', date)[0]
            sns.lineplot(x='time_hours', y='p_VC_mbar', data=data, ax=ax, label=date, marker='o')

        ax.set(yscale="log")
        ax.set_ylabel('p_VC_mbar [mbar]')


        ax.set_xlabel(f'Hour')

        ax.set_ylim(1e-7, 1e-3)

        plt.show()
```

```python
pdc = pumping_down_graph()

fname1 = '/Users/hkromer/02_PhD/02_Data/11.pumping_down/2019-11-29.txt'
fname2 = '/Users/hkromer/02_PhD/02_Data/11.pumping_down/2019-12-03.txt'
fname3 = '/Users/hkromer/02_PhD/02_Data/11.pumping_down/2019-12-17.txt'

pdc.plot_graph_VC(fname1, fname2, fname3)
```
