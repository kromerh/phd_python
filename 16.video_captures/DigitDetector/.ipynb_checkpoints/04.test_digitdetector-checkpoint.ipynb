## 1. Load the DigitDetector model

```python
path = '/Users/hkromer/02_PhD/02_Data/12.dcr/Stanford_housenumbers/2019-19-25.KERAS_model_DigitDetector.h5'
svhn = SVHNDataset()

model = svhn.load_model(path)
model.summary()
```

```python
# test on some examples
cred = '/Users/hkromer/02_PhD/tmp/cred_camera.pw'
pro = PressureReadoutObject(cred)
img = pro.read_test_image('/Users/hkromer/02_PhD/02_Data/12.dcr/data_f_gray.csv')
# plt.imshow(img, cmap='gray')
# plt.show()


# convert
img = img.astype(np.float64)
# loop through the image in steps of 2 pixels with framesize and check if it is a digit in the frame or not
def digit_in_frame(model, frame):
    """
    Determine if a digit is in the frame or not.
    INPUT:
        - model: KERAS model trained to classify digit or not.
        - frame: the frame to be tested.
    OUTPUT:
        - returns 1 if there is a digit in the frame, otherwise 0
    """
    assert frame.shape == (1,32,32,1)
    y_predict = model.predict(frame)
    return y_predict

img = pro.scale_image(img,scale_perc=45)
framesize = 32 # framesize, take 32
width = img.shape[1]
height = img.shape[0]
stepsize = 8 # how many pixels to move the frame
frames_with_image = [[],[],[]] # ii, jj, probability
threshold = 0.95 # for prediction of positive class
l_horizontal = np.arange(0,width-framesize+stepsize,stepsize)
l_vertical = np.arange(0,height-framesize+stepsize,stepsize)
combinations = len(l_horizontal) * len(l_vertical)
print(f'Doing {combinations} combinations')
count = 0
for ii in l_horizontal: # horizontal
    print(f'Doing combination {count} out of {combinations}')
    for jj in l_vertical:# vertical
        count = count + 1
        frame = img[jj:jj+framesize,ii:ii+framesize]
        frame = frame.reshape(-1,32,32,1)
        y_predict = digit_in_frame(model, frame)[0]
#         print(ii,jj,y_predict)
        if y_predict[1] > threshold:
            frames_with_image[0].append(ii)
            frames_with_image[1].append(jj)
            frames_with_image[2].append(y_predict)
        
```

```python
# re-define the threshold
threshold = 0.95 # for prediction of positive class
new_l = [[],[],[]]
for ii in range(0,len(frames_with_image[0])):
    preds = frames_with_image[2][ii] # predictions            
    if preds[1] >= threshold:
        new_l[0].append(frames_with_image[0][ii]) # ii
        new_l[1].append(frames_with_image[1][ii]) # jj
        new_l[2].append(frames_with_image[2][ii]) # pred

# visualize them in the image
iis = new_l[0] # horizontal
jjs = new_l[1] # vertical
preds = new_l[2] # predictions
```

```python
print(f'Found {len(frames_with_image[0])} possible digits in the image.')         

iis = frames_with_image[0] # horizontal
jjs = frames_with_image[1] # vertical
preds = frames_with_image[2] # predictions



# select 10 at random
idx = [np.random.randint(0,len(iis)) for ii in range(0,10)]

def draw_rectangle(img, size, p0, p1, border=5):
    """
    Draws a rectangle with bordersize in the image img at position p0 (vertical), p1 (horizontal). The framesize of the sub-image is size.
    returns the image with the rectangle
    """
    img_t = img.copy()
    img_t[p0:p0+size+border,p1:p1+border] = 255.0 # left side
    img_t[p0:p0+size+border,p1+size:p1+size+border] = 255.0 # right side
    img_t[p0:p0+border,p1:p1+size] = 255.0 # top side
    img_t[p0+size:p0+size+border,p1:p1+size] = 255.0 # bottom side
    
    return img_t


for kk in range(0,len(jjs)):
    img_0 = draw_rectangle(img, size=32, p0=jjs[kk], p1=iis[kk], border=5)
    # draw a border around the image
    plt.imshow(img_0)
    prob = int(round(preds[kk][1]*100))
    print(f'Probability: {prob}')
    plt.text(iis[kk]+8, jjs[kk]-5, f'{prob}', color='red', weight = 'heavy')
    plt.show()
```

```python
# ground truth images
p0 = 82
p1 = 10
img_0 = draw_rectangle(img, 32, p0, p1, border=1)
plt.imshow(img_0)
plt.show()

p0 = 82
p1 = 31
img_0 = draw_rectangle(img, 32, p0, p1, border=1)
plt.imshow(img_0)
plt.show()

p0 = 82
p1 = 80
img_0 = draw_rectangle(img, 32, p0, p1, border=1)
plt.imshow(img_0)
plt.show()
```
