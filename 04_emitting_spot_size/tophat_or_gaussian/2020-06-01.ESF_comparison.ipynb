# Comparison of the ESF of a gaussian and tophat defined source from MCNP

- The question is how does the ESF look like in either case, can the shape be used to determine if it is a Gaussian or more Tophat souce?

- This is for the 5x5mm2 detector, new chamber, gauss and tophat definition

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the two dataframes. They have the same settings, but different sizes in the MCNP source definition (gaussian profile)
dfs = []
for ii in [2,3]:
    
    fname = f'/Users/hkromer/02_PhD/02_Data/09.NeutronEmittingSpot/02.Experiment/lookup_table_simulation/df_MCNP_results_0{ii}.csv'
    _ = pd.read_csv(fname, index_col=0)
#     print(np.min(_['x_pos'].unique()),np.max(_['x_pos'].unique()) )
    dfs.append(_)    
data_gauss = pd.concat(dfs, ignore_index=True)
data_gauss = data_gauss.drop_duplicates() # 6 mm occurs in both datasets
data_gauss.head()
```

```python
data_gauss['radius'].unique() # radius is the fwhm in cm
```

```python
data_tophat = pd.read_csv('/Users/hkromer/02_PhD/tmp/FINAL/df_MCNP_results.csv', index_col=0)
data_tophat.head()
```

## Gaussian and tophat closest value

For 2.5 mm FWHM seen by detector (see 02_PhD/01.github/phd/04_emitting_spot_size/03.Lookup_table_simulation_MCNP6.ipynb)

```python
# closest value for 2.5 mm FWHM seen by detector is 2.95 mm definition as radius
val_gauss = 0.295 # fwhm in cm
# val_gauss = 0.17 # fwhm in cm
val_tophat = 4.2 # diameter in mm
# val_tophat = 2.7 # diameter in mm

data_gauss_1 = data_gauss.loc[ data_gauss['radius'] == val_gauss ]
data_gauss_1 = data_gauss_1.sort_values(by='x_pos')
data_gauss_1.head()
```

```python
data_tophat_1 = data_tophat.loc[ data_tophat['diameter'] == val_tophat ]
data_tophat_1 = data_tophat_1.sort_values(by='x_pos')
data_tophat_1.head()
```

## Normalize the counts

```python
vals_norm = 10
norm_cps = data_gauss_1['Cps_cutoff'].values[:-vals_norm].mean()
data_gauss_1['Cps_cutoff'] = data_gauss_1['Cps_cutoff'].values / norm_cps

norm_cps = data_tophat_1['Cps_cutoff'].values[:-vals_norm].mean()
data_tophat_1['Cps_cutoff'] = data_tophat_1['Cps_cutoff'].values / norm_cps

scale_cps = data_gauss_1['Cps_cutoff'].values[0:vals_norm].mean()
data_gauss_1['Cps_cutoff'] = data_gauss_1['Cps_cutoff'].values - scale_cps

scale_cps = data_tophat_1['Cps_cutoff'].values[0:vals_norm].mean()
data_tophat_1['Cps_cutoff'] = data_tophat_1['Cps_cutoff'].values - scale_cps
```

```python
%matplotlib widget

# HEAD
sns.set(context='paper', font_scale=1.5)
fig, ax = plt.subplots(figsize=(8,6))

# GAUSS
X = data_gauss_1['x_pos'].values
Y = data_gauss_1['Cps_cutoff'].values

plt.plot(X, Y, marker='o', markersize=2, color='darkblue', label='Gauss')


# TOPHAT
X = data_tophat_1['x_pos'].values
Y = data_tophat_1['Cps_cutoff'].values

plt.plot(X, Y, marker='o', markersize=2, color='darkorange', label='Tophat')


# LAYOUT
plt.ylabel('Simulated cps in detector (normalized) [-]')
plt.xlabel('Edge position [mm]')
plt.legend()
plt.tight_layout()
plt.show()
```

### Entering tracks

```python
%matplotlib widget

vals_norm = 10
norm_cps = data_gauss_1['entering_tracks'].values[:-vals_norm].mean()
data_gauss_1['entering_tracks'] = data_gauss_1['entering_tracks'].values / norm_cps

norm_cps = data_tophat_1['entering_tracks'].values[:-vals_norm].mean()
data_tophat_1['entering_tracks'] = data_tophat_1['entering_tracks'].values / norm_cps

scale_cps = data_gauss_1['entering_tracks'].values[0:vals_norm].mean()
data_gauss_1['entering_tracks'] = data_gauss_1['entering_tracks'].values - scale_cps

scale_cps = data_tophat_1['entering_tracks'].values[0:vals_norm].mean()
data_tophat_1['entering_tracks'] = data_tophat_1['entering_tracks'].values - scale_cps

# HEAD
sns.set(context='paper', font_scale=1.5)
fig, ax = plt.subplots(figsize=(8,6))

# GAUSS
X = data_gauss_1['x_pos'].values
Y = data_gauss_1['entering_tracks'].values

plt.plot(X, Y, marker='o', markersize=2, color='darkblue', label='Gauss')


# TOPHAT
X = data_tophat_1['x_pos'].values
Y = data_tophat_1['entering_tracks'].values

plt.plot(X, Y, marker='o', markersize=2, color='darkorange', label='Tophat')


# LAYOUT
plt.ylabel('Simulated entering tracks in detector (normalized) [-]')
plt.xlabel('Edge position [mm]')
plt.legend()
plt.tight_layout()
plt.show()
```

### R tot


```python
%matplotlib widget

vals_norm = 10
norm_cps = data_gauss_1['R_tot'].values[:-vals_norm].mean()
data_gauss_1['R_tot'] = data_gauss_1['R_tot'].values / norm_cps

norm_cps = data_tophat_1['R_tot'].values[:-vals_norm].mean()
data_tophat_1['R_tot'] = data_tophat_1['R_tot'].values / norm_cps

scale_cps = data_gauss_1['R_tot'].values[0:vals_norm].mean()
data_gauss_1['R_tot'] = data_gauss_1['R_tot'].values - scale_cps

scale_cps = data_tophat_1['R_tot'].values[0:vals_norm].mean()
data_tophat_1['R_tot'] = data_tophat_1['R_tot'].values - scale_cps

# HEAD
sns.set(context='paper', font_scale=1.5)
fig, ax = plt.subplots(figsize=(8,6))

# GAUSS
X = data_gauss_1['x_pos'].values
Y = data_gauss_1['R_tot'].values

plt.plot(X, Y, marker='o', markersize=2, color='darkblue', label='Gauss')


# TOPHAT
X = data_tophat_1['x_pos'].values
Y = data_tophat_1['R_tot'].values

plt.plot(X, Y, marker='o', markersize=2, color='darkorange', label='Tophat')


# LAYOUT
plt.ylabel('Simulated R_tot in detector (normalized) [-]')
plt.xlabel('Edge position [mm]')
plt.legend()
plt.tight_layout()
plt.show()
```
